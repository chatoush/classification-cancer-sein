# 1. Importation des modules

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (precision_score, recall_score, accuracy_score, f1_score, classification_report, confusion_matrix)
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, cross_validate
```

# 2. Importation des données

```python
df = load_breast_cancer(as_frame = True)
data = df.frame
data.head()
```

# 5. Préparation des données 

```python
y = data["target"]
X = data.drop("target", axis = 1)
```

On sépare le tableau data en deux, X contient les features et y le target

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20,random_state = 42, stratify = y)
```

On sépare les données en données d’entrainement (X_train, y_train) et en données test (x_test, y_test) en fixant les parts et divisant y proportionnellement 

# 6. Pipeline et validation croisée

```python
model = Pipeline([
    ("scaling", StandardScaler()),
    ("Classification", LogisticRegression())
])
```

On crée un pipeline pour automatiser la normalisation et l'application du model choisi, ici, la classification

```python
cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)
```

```python
scores = cross_validate(
    model,
    X,
    y,
    cv=cv,
    scoring=["accuracy", "precision", "recall", "f1"],
    n_jobs=-1
)
```

```python
print("Accuracy moyenne :", scores["test_accuracy"].mean())
print("Precision moyenne :", scores["test_precision"].mean())
print("Recall moyen :", scores["test_recall"].mean())
print("F1-score moyen :", scores["test_f1"].mean())
```

Les résultats de la validation croisée montrent que le modèle est globalement stable et performant sur différents sous-ensembles des données.
Le modèle obtient une accuracy moyenne d’environ 97 %, ainsi qu’un excellent recall (99 %), ce qui indique qu’il détecte efficacement les cas positifs avec très peu de faux négatifs.
Les scores de précision et de F1-score élevés montrent également un bon équilibre global des performances.

# 7. Évaluation du modèle

```python
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test,y_predict)
precision = precision_score(y_test,y_predict)
recall = recall_score(y_test,y_predict)
f1 = f1_score(y_test,y_predict)


print("Accurary  : ", round(accuracy, 3))
print("Precision : ", round(precision, 3))
print("Recall    : ", round(recall, 3))
print("F1-score  : ", round(f1, 3))
```

```python
print("\n Rapport de classifcation :")
print(classification_report(y_test,y_predict))
```

```python
cm = confusion_matrix(y_test,y_predict)

plt.figure(figsize=(6,4))
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=['Prédict: Maligne', 'Predict: Bénigne'],
    yticklabels=['Réel: Maligne', 'Réel: Bénigne']
)

plt.title("Matrice de confusion")
plt.xlabel("Prédiction")
plt.ylabel("Valeur réelle")
plt.show()
```

Les résultats finaux montrent que le modèle possède d’excellentes performances de classification.
La matrice de confusion indique que la majorité des observations sont correctement classées, avec seulement deux erreurs de prédiction.
Les métriques élevées (accuracy, precision, recall et F1-score) confirment la capacité du modèle à distinguer efficacement les tumeurs bénignes et malignes.

La régression logistique obtient de meilleurs résultats que Random Forest sur ce dataset. Cela peut s’expliquer par le fait que les données sont relativement bien séparées et structurées, ce qui favorise les modèles linéaires simples.
